```python
import matplotlib.pyplot as plt
import numpy as np
```

### Lesson 1

```python
# line chart
```

```python
# data for x-axis
x = np.linspace(0, 10, 100)

# data for y-axis
y = np.sin(x)
```

```python
plt.plot(x,y)
plt.xlabel('time')
plt.ylabel('function')
plt.title('sin wave')
plt.show()
```
